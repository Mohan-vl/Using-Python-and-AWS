New lambda function created referring zip file, comprising lambda handler and boto3 package.
Function detect_entities29, uses Amazon Comprehend service and comprehend.detect_entities method to identify entities/organization referred in the text file


```python
!aws lambda create-function  \
--function-name detect_entities29  \
--runtime python3.7  \
--zip-file fileb://detect_entities.zip  \
--handler graded_lambda.lambda_handler  \
--role arn:aws:iam::808392735438:role/csdlambdarole  \
--region us-east-1
```

```python
lam = boto3.client ('lambda',region_name='us-east-1')
```

```python
response = lam.invoke(
FunctionName = 'detect_entities29',
InvocationType ='RequestResponse',
LogType  = 'Tail',
  
Payload = json.dumps({"bucket":"testbuck-3","object_name":"Graded_Review_Analysis.txt"})  
)

entities = response['Payload'].read().decode('utf8')
```

- Function to detect entities has been invoked and results are stored in entities (list)

```python
print (entities)
finalentities = json.loads (entities)
```

```python
for objects in finalentities:
     if objects['Type'] == "ORGANIZATION":
            print(objects['Text'])
```

In the above statement, All applicable organization entities are being listed
