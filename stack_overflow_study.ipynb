```python
ques_eta_df['CreationDate'] = pd.to_datetime(ques_eta_df['CreationDate'], format = "%Y-%m-%dT%H:%M:%SZ")
ques_eta_df['ClosedDate'] = pd.to_datetime(ques_eta_df['ClosedDate'], format = "%Y-%m-%dT%H:%M:%SZ")
```

```python
ans_eta_df = ans_df[ans_df['Score'] > 0]
print(ans_eta_df.isnull().sum())
```

```python
ans_eta_df['CreationDate'] = pd.to_datetime(ans_eta_df['CreationDate'], format = "%Y-%m-%dT%H:%M:%SZ")
ans_eta_df.Score.describe()
```

```python
ans_eta_firstans_df = ans_eta_df.groupby(['ParentId'])['CreationDate'].min();
```

```python
ans_eta_firstans_df.name = 'AnsMinCreationDt';
```

```python
ans_eta_firstans_df.head(n=1000)
```

 *ans_eta_firsttrans_df is selection of answered questions with Min, earliest answered*

```python
ques_eta_df = ques_eta_df.merge (ans_eta_firstans_df, how='inner',left_on='Id',right_on='ParentId')
ques_eta_df.head(n=1000)
```

```python
ques_eta_df['Answered_Time'] = ques_eta_df['AnsMinCreationDt'] - ques_eta_df['CreationDate']
ques_eta_df.head(n=1000)
```

```python
ques_eta_df = ques_eta_df[ques_eta_df['Answered_Time'] > '0 days 00:00:00']
ques_eta_df.describe()
```

```python
print(tags_df.head(n=1000))
ques_eta_df.head(n=1000)
```

```python
ques_eta_df = ques_eta_df.merge (tags_df, how='inner',left_on='Id',right_on='Id')
ques_eta_df['Answered_Time_Int'] = ques_eta_df['Answered_Time'].values.astype(np.int64)
ques_eta_df.head(n=1000)

```

```python
datascience = ['python', 'r', 'matlab', 'sas', 'excel', 'sql'] 
# selecting rows based on condition 
ques_eta_df = ques_eta_df[ques_eta_df['Tag'].isin (datascience)]
```

```python
ques_eta_df.head(n=1000)

```

```python
#eta_first_ans_df = ques_eta_df.groupby('Tag').mean();
#eta_first_ans_df = pd.DataFrame(ques_eta_df.groupby('Tag').mean());
eta_first_ans_df = ques_eta_df.groupby(['Tag'])['Answered_Time_Int'].mean();
eta_first_ans_df.head(n=1000)  
eta_first_answer_df = pd.DataFrame({'Tag':eta_first_ans_df.index, 'Answered_Time_Int':eta_first_ans_df.values})
eta_first_answer_df.head(n=1000)  
eta_first_answer_df['Answered_Time'] = pd.to_timedelta(eta_first_answer_df['Answered_Time_Int'])
eta_first_answer_df.head(n=1000)  
```
